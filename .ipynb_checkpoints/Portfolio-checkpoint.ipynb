```python
from alpha_vantage.timeseries import TimeSeries

API_key = 'OZHBQ2Q48QC0NFRZ'
ts = TimeSeries(key = API_key,output_format='pandas')

def stock_pull(stock, startdate, enddate, interval=None):
    if interval==None:
        data = ts.get_daily_adjusted(stock, outputsize='full')
        data = data[0].reset_index()
        data = data[(data['date'] >= startdate) & (data['date'] <= enddate)]
        data = data.sort_values('date', ascending=True)
        data['percent change'] = data['4. close'].pct_change()
    else:
        data = ts.get_intraday(stock, interval = interval, outputsize='full')
        data = data[0].reset_index()
        data = data[(data['date'] >= startdate) & (data['date'] <= enddate)]
        data=data.sort_values('date', ascending=True)
        data['percent change'] = data['4. close'].pct_change()
    return data

```

```python
### Take the closing values of each stock and put them into a new dataframe ###
### You'll have to insert them manually for now ###

stonks = pd.DataFrame([AAPL.iloc[:,5],MSFT.iloc[:,5], AMZN.iloc[:,5], JD.iloc[:,5], VGT.iloc[:,5], 
          BABA.iloc[:,5], KO.iloc[:,5], GOOG.iloc[:,5], T.iloc[:,5], EA.iloc[:,5], NDAQ.iloc[:,5], 
          NUEM.iloc[:,5], EPD.iloc[:,5], NEE.iloc[:,5], PYPL.iloc[:,5]], 
                      index = ['AAPL', 'MSFT', 'AMZN', 'JD', 'VGT','BABA','KO','GOOG','T',
                                 'EA','NDAQ','NUEM','EPD','NEE','PYPL']).transpose()
### Set index to datetime using one of the stocks above ###
ind = AAPL['date']
stonks['date'] = ind
table = stonks.set_index('date')
table = table.iloc[::-1]
```

```python
### enter the dataframe, number of portfolios to test, and current riskfree rate ###
### The US 10-year treasury bond is usually used as the riskfree rate ###
def ex_return_weighted(stonks, num_portfolios=10000, riskfree_rate=.0162):
    results = np.zeros((stonks.shape[1],num_portfolios))
    weights_record = []
    for i in range(num_portfolios):
        weights = np.random.random(stonks.shape[1])
        weights /= np.sum(weights)
        weights_record.append(weights)
    weights_record = np.array(weights_record)
    stonks_growth = np.array(stonks.pct_change().mean())

    daily_expected_return = (weights_record*stonks_growth)
    expected_return = daily_expected_return.sum(axis = 1)*252
    std_record = []
    for weight in weights_record:
        std_record.append(np.nanstd(np.array(stonks.pct_change()*252) *np.array(weight)))
    std_record = np.array(std_record)
    
    sharpe_ratio = (expected_return-riskfree_rate)/std_record
    
    return weights_record, expected_return, std_record, sharpe_ratio

### returned are the randomly weighted portfolios, expected return and standard deviation for both ###
### and the sharpe ratio which is used for "maximizing" portfolio returns. ###


ex = ex_return_weighted(table)
ex
```

```python
fig,ax = plt.subplots()
### plots the portfolios and captures the "Ideal" portfolio balance. ###
### i.e. to maximize returns I would invest 5.85% of my portfolio into AAPL ###
def efficient_frontier(stonks, num_portfolios, riskfree_rate, ax):
    weights, ex_return, std_port, sharpe_ratio = ex_return_weighted(stonks, num_portfolios, riskfree_rate)
    ax.scatter(std_port, ex_return)
    best_port = np.argmax(sharpe_ratio)
    ax.scatter(std_port[best_port], ex_return[best_port], color = 'red')
    ax.set_xlabel('Risk')
    ax.set_ylabel('Expected Return')
    return pd.DataFrame([weights[best_port]], index = ['weight'], columns = stonks.columns)
    
efficient_frontier(table, 10000, .005, ax)
```
